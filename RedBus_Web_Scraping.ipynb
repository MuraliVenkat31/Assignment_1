## kerala

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Open the browser
driver = webdriver.Chrome()

# Load the webpage
driver.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")

# Maximize the window
driver.maximize_window()

# Wait for the elements to load
wait = WebDriverWait(driver, 20)

# Function to scrape route links and names
def scrape_routes():
    route_links = []
    route_names = []

    # Loop through pages
    for i in range(1, 3):
        try:
            # Wait for the routes to load
            routes = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[class ="route"]')))

            # Extract route links and names
            for route in routes:
                route_links.append(route.get_attribute("href"))
                route_names.append(route.text)

            # Wait for the pagination element to be present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')

            # Click the next button
            driver.execute_script("arguments[0].click();", next_button)

        except Exception as e:
            print(f"Error: {e}")
            break

    return route_links, route_names

# Scrape routes
route_links, route_names = scrape_routes()      # Calls the scrape_routes function and stores the returned data in route_links and route_names.

# Create a DataFrame
df = pd.DataFrame({"Route Name": route_names, "Route Link": route_links})

# Print the DataFrame
print(df)

# Close the browser
driver.quit()
```

```python
df.to_csv("Bus_Routes.csv",index = False)
```
